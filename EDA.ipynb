## Reemplazar valores faltantes 

Reemplazamos algunos valores nulos usando informacion de titulos que se 
repiten en 2 o mas data frames

```python
def remplazar_cast(df1, df2):
    """
    Remplaza los valores de la columna 'cast' del data frama df1 que aparecen 
    como 'No Info' buscandolos en el data frame df2, esta busqueda se hace teniendo 
    en cuenta la columna 'title'
    """
    
    df1_null = df1[['title','release_year']][df1['cast'] == 'No Info']
    
    for index in  df1_null.index:
        if df2['title'].loc[(df2['title'] == df1_null['title'].loc[index])].shape[0] != 0:
            df1['cast'].loc[index] = df2['cast'].loc[df2['title'] == 
                                                     df1_null['title'].loc[index]].iloc[0]  
            
    return df1
```

```python
for i, item in enumerate(data.cast):
    if item.find('/') != -1:
        data.cast.iloc[i] = item.replace('/',',')
        
for i, item in enumerate(data.cast):
    if item.find('|') != -1:
        data.cast.iloc[i] = item.replace('|',',')
```

Eliminar ', Jr.' y , Ph.D.

```python
for i, item in enumerate(data.cast):
    if item.find(', Jr.') != -1:
        data.cast.iloc[i] = item.replace(', Jr.','')
        
for i, item in enumerate(data.cast):
    if item.find(', Ph.D.') != -1:
        data.cast.iloc[i] = item.replace(', Ph.D.','')
```

```python
for item in data.cast:
    if item.find(', Jr.') != -1:
        print(item)
```

Eliminamos todo tipo de acentos

```python
def eliminar_acentos(cad):
    s = cad
    # -> NFD y eliminar diacr√≠ticos
    s = re.sub(
        r"([^n\u0300-\u036f]|n(?!\u0303(?![\u0300-\u036f])))[\u0300-\u036f]+", r"\1", 
        normalize( "NFD", s), 0, re.I
    )

    # -> NFC
    s = normalize( 'NFC', s)
    return s
```

```python
for i, item in enumerate(data.cast):
    data.cast.iloc[i] = eliminar_acentos(item)
```

Eliminar comillas simples y dobles

```python
for i, item in enumerate(data.cast):
    if item.find("'") != -1:
        item = item.replace("'","")
        data.cast.iloc[i] = item
    if item.find('"') != -1:
        item = item.replace('"','')
        data.cast.iloc[i] = item
```

## Creamos tabla Movie



```python
Movie = pd.DataFrame(columns=['IdMovie', 'Title', 'IdType','IdPlatform', 
                              'ReleaseYear', 'Duration', 'TimeUnit'])
Movie
```

Eliminamos comillas simples y dobles de la columna 'title'

```python
for i, item in enumerate(data.title):
    if item.find("'") != -1:
        item = item.replace("'","")
        data.title.iloc[i] = item
    if item.find('"') != -1:
        item = item.replace('"','')
        data.title.iloc[i] = item
```

```python
for i in range(data.shape[0]):
    
    if data.type.iloc[i] == 'Movie':
        type_ = 1 
    elif data.type.iloc[i] == 'TV Show':
        type_ = 2
        
    if data.platform.iloc[i] == 'Amazon':
        platform = 1
    elif data.platform.iloc[i] == 'Disney':
        platform = 2
    elif data.platform.iloc[i] == 'Hulu':
        platform = 3
    elif data.platform.iloc[i] == 'Netflix':
        platform = 4
        
    aux = data.duration.iloc[i].split(' ')
    
    if aux[0] == 'No' and aux[1] == 'Info':
        duration = np.nan
        time_unit = 'No Info'
    else:
        duration = int(aux[0])
        if aux[1] == 'min':
            time_unit = 'min'
        elif aux[1] == 'Season' or aux[1] == 'Seasons':
            time_unit = 'season'
        
    Movie = Movie.append({'IdMovie': i+1, 
                          'Title': data.title.iloc[i], 
                          'IdType': type_,
                          'IdPlatform': platform, 
                          'ReleaseYear': data.release_year.iloc[i], 
                          'Duration': duration, 
                          'TimeUnit': time_unit},
                         ignore_index=True)
    
```

```python
Movie.to_csv('./Tablas/Movie.csv', index=False, encoding='utf-8')
```
